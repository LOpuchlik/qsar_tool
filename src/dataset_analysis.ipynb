### Groups of molecules are read using a Supplier
(for example, an rdkit.Chem.rdmolfiles.SDMolSupplier
or a rdkit.Chem.rdmolfiles.SmilesMolSupplier)

```python
supplier = Chem.SDMolSupplier('../data/input/example_5ht3ligs.sdf')
```

```python
supplier
```

### Displaying molecules

```python
for molecule in supplier:
    img = Chem.Draw.MolToImage(molecule)
    img.show() ## shows in a separate window
```

```python
for i, molecule in enumerate(supplier):
    img = Chem.Draw.MolToImage(molecule)
    display(img) ## shows inline in notebook
    ## img.save(f"../data/output/molecule_{i+1}.png") ## saving image
```
